```python
from pylab import *
%matplotlib inline 
```

```python
def RK4(funcion,p_ini,x,h=0.1):
    tiempos=arange(x[0],x[1]+h,h)
    sol = zeros((len(tiempos),len(p_ini)))
    sol[0,:]=p_ini
    for i in xrange(len(tiempos)-1):
        t=tiempos[i]
        y_n=sol[i,:]
        k_1=funcion(y_n,t)
        k_2=funcion(y_n + 0.5*h*k_1,t+0.5*h)
        k_3=funcion(y_n + 0.5*h*k_2,t+0.5*h)
        k_4=funcion(y_n + k_3,t+h)
        sol[i+1,:]= y_n + (h/6)*(k_1+2*k_2+2*k_3+k_4)
    return sol,tiempos
```

```python
def ec_dif(y,t):
    return array([y[1],-y[0]])

y,t=RK4(ec_dif,[1,0],[0,2*pi])

figure(figsize=(16,6),dpi=100)
subplot(1,2,1)
plot(t,y[:,0],t,y[:,1])
subplot(1,2,2)
plot(y[:,0],y[:,1])
```
