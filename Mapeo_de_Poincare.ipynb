```python
def osc(y,t,w02=1.):
    return array([y[1],-w02*w02*sin(y[0])+0.1*sin(t)])
```

```python
figure(figsize=(10, 10), dpi=80)
for i in xrange(2):
    y0,u0= 1.2 + 0.01*i,0
    sol,t=RK4(osc,[y0,u0],[0,10000.])
    plot(sol[:,0],sol[:,1])
axes().set_aspect( 'equal' ),grid(True)
xlim([-pi,pi]),ylim([-pi,pi])
show()
```
