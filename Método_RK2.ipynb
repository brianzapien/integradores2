```python
from pylab import *
%matplotlib inline 
```

```python
def RK2(funcion,p_ini,x,h=0.1):
    tiempos=arange(x[0],x[1]+h,h)
    sol = zeros((len(tiempos),len(p_ini)))
    sol[0,:]=p_ini
    for i in xrange(len(tiempos)-1):
        t=tiempos[i+1]
        y_n=sol[i,:]
        k_1=h*funcion(y_n,t)
        sol[i+1,:]= y_n + h*funcion(y_n+k_1,t+0.5*h)
    return sol,tiempos
```

```python
def ec_dif(y,t):
    return array([y[1],-2*y[1]-0.5*sin(y[0])+sin(t)])

y,t=RK2(ec_dif,[1,0],[0,20*pi],0.01)

figure(figsize=(16,6),dpi=100)
subplot(1,2,1)
plot(t,y[:,0],t,y[:,1])
subplot(1,2,2)
plot(y[:,0],y[:,1])
```
