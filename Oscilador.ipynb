```python
from pylab import *
from scipy.integrate import odeint
from JSAnimation import IPython_display
from matplotlib import animation
%matplotlib inline
```

```python
def osc(y_v,t,w=1.,mu=0):
    dy=y_v[1]
    dvy=-w*w*y_v[0]-mu*dy
    return array([dy,dvy])
```

```python
x=linspace(0,100,500)
sol=odeint(osc,[0,1],x,args=(1,0.1))
plot(sol[:,0],sol[:,1])
```

```python
figura=figure()
ejes= axes(xlim=(-1,1),ylim=(-1,1))

trayectoria, = ejes.plot([],[],lw=2)
punto, = ejes.plot([],[],lw=2.5,'o')

def init():
    trayectoria.set_data([],[])
    punto.set_data([],[])
    return trayectoria,punto,

def el_movimiento(i):
    y,v=sol[:,0],sol[:,1]
    trayectoria.set_data(y[:i+1],v[:i+1])
    punto.set_data(y[i],v[i])
    return trayectoria,punto,

animation.FuncAnimation(figura,el_movimiento,init_func=init,frames=len(sol[:,0]))
```
